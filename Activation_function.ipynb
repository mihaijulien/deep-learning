```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_activation_functions(fcn):
  x_vals = np.linspace(-3,3,100)  # afisam valorile intre intervalul -3 3 si definim 100 de puncte in acest interval

  y_vals = fcn(x_vals)

  plt.plot(x_vals, y_vals)
  plt.xlabel("Valoare input")
  plt.ylabel("Valoare output")

  plt.title("Vizualizare functie de activare")
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
visualize_activation_functions(sigmoid)
```

```python
def tanh_np(x):
    return np.tanh(x)

def tanh(x):
    return (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
```

```python
visualize_activation_functions(tanh)
```

```python
# Tanh vs Sigmoid

plt.figure()
visualize_activation_functions(tanh)
visualize_activation_functions(sigmoid)
plt.show()
```
